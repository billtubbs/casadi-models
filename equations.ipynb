## Equations for state space model derived from transfer function

From den = [1 -2.3 2.6 -1.8 0.72 -0.12], we have:
  - $a_1 = -2.3, a_2 = 2.6, a_3 = -1.8, a_4 = 0.72, a_5 = -0.12$

From num = [0.01 0.03 0.03 0.02 0.01] (with implicit delay), we have:
  - $b_1 = 0.01, b_2 = 0.03, b_3 = 0.03, b_4 = 0.02, b_5 = 0.01$

The state-space model in observable canonical form is:

$$\mathbf{x}(k+1) = \begin{bmatrix}
0 & 0 & 0 & 0 & -a_5 \\
1 & 0 & 0 & 0 & -a_4 \\
0 & 1 & 0 & 0 & -a_3 \\
0 & 0 & 1 & 0 & -a_2 \\
0 & 0 & 0 & 1 & -a_1 \\
\end{bmatrix} \mathbf{x}(k) + \begin{bmatrix}
b_5 \\ b_4 \\ b_3 \\ b_2 \\ b_1
\end{bmatrix} u(k)$$

$$y(k) = \begin{bmatrix} 0 & 0 & 0 & 0 & 1 \end{bmatrix} \mathbf{x}(k)$$

Substituting the values:

$$\mathbf{x}(k+1) = \begin{bmatrix}
0 & 0 & 0 & 0 & 0.12 \\
1 & 0 & 0 & 0 & -0.72 \\
0 & 1 & 0 & 0 & 1.8 \\
0 & 0 & 1 & 0 & -2.6 \\
0 & 0 & 0 & 1 & 2.3 \\
\end{bmatrix} \mathbf{x}(k) + \begin{bmatrix}
0.01 \\ 0.02 \\ 0.03 \\ 0.03 \\ 0.01
\end{bmatrix} u(k)$$

**General form for transfer function to observable canonical state-space:**

Given a discrete-time transfer function:

$$G(z) = \frac{b_0 z^n + b_1 z^{n-1} + \cdots + b_n}{z^n + a_1 z^{n-1} + \cdots + a_n}$$

The observable canonical form state-space representation is:

$$\mathbf{x}(k+1) = \begin{bmatrix}
0 & 0 & \cdots & 0 & -a_n \\
1 & 0 & \cdots & 0 & -a_{n-1} \\
0 & 1 & \cdots & 0 & -a_{n-2} \\
\vdots & \vdots & \ddots & \vdots & \vdots \\
0 & 0 & \cdots & 1 & -a_1 \\
\end{bmatrix} \mathbf{x}(k) + \begin{bmatrix}
b_n \\ b_{n-1} \\ b_{n-2} \\ \vdots \\ b_1
\end{bmatrix} u(k)$$

$$y(k) = \begin{bmatrix} 0 & 0 & \cdots & 0 & 1 \end{bmatrix} \mathbf{x}(k) + b_0 \, u(k)$$

where $D = b_0$ (typically zero for proper systems with delay).
