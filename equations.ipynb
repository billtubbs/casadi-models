## Equations for state space model derived from transfer function

The `StateSpaceModelDTTFSISO` class implements a discrete-time transfer function model:

$$G(z) = \frac{B(z)}{A(z)} = \frac{b_0 z^n + b_1 z^{n-1} + \cdots + b_n}{z^n + a_1 z^{n-1} + \cdots + a_n}$$

where $z$ is the forward shift operator in the z-transform domain.

This translates to the following difference equation:

$$y(k) = -a_1 y(k-1) - a_2 y(k-2) - \cdots - a_n y(k-n) + b_0 u(k) + b_1 u(k-1) + \cdots + b_n u(k-n)$$

The model is internally converted to state-space form using the observable canonical form representation.

Example:

From `den = [1 -2.3 2.6 -1.8 0.72 -0.12]`, we have:
  - $a_1 = -2.3, a_2 = 2.6, a_3 = -1.8, a_4 = 0.72, a_5 = -0.12$

From `num = [0.01 0.03 0.03 0.02 0.01]` (with implicit delay), we have:
  - $b_1 = 0.01, b_2 = 0.03, b_3 = 0.03, b_4 = 0.02, b_5 = 0.01$

The state-space model in observable canonical form is:

$$\mathbf{x}(k+1) = \begin{bmatrix}
0 & 0 & 0 & 0 & -a_5 \\
1 & 0 & 0 & 0 & -a_4 \\
0 & 1 & 0 & 0 & -a_3 \\
0 & 0 & 1 & 0 & -a_2 \\
0 & 0 & 0 & 1 & -a_1 \\
\end{bmatrix} \mathbf{x}(k) + \begin{bmatrix}
b_5 \\ b_4 \\ b_3 \\ b_2 \\ b_1
\end{bmatrix} u(k)$$

$$y(k) = \begin{bmatrix} 0 & 0 & 0 & 0 & 1 \end{bmatrix} \mathbf{x}(k)$$

Substituting the values:

$$\mathbf{x}(k+1) = \begin{bmatrix}
0 & 0 & 0 & 0 & 0.12 \\
1 & 0 & 0 & 0 & -0.72 \\
0 & 1 & 0 & 0 & 1.8 \\
0 & 0 & 1 & 0 & -2.6 \\
0 & 0 & 0 & 1 & 2.3 \\
\end{bmatrix} \mathbf{x}(k) + \begin{bmatrix}
0.01 \\ 0.02 \\ 0.03 \\ 0.03 \\ 0.01
\end{bmatrix} u(k)$$

**General form for transfer function to observable canonical state-space:**

Given a discrete-time transfer function:

$$G(z) = \frac{b_0 z^n + b_1 z^{n-1} + \cdots + b_n}{z^n + a_1 z^{n-1} + \cdots + a_n}$$

The observable canonical form state-space representation is:

$$\mathbf{x}(k+1) = \begin{bmatrix}
0 & 0 & \cdots & 0 & -a_n \\
1 & 0 & \cdots & 0 & -a_{n-1} \\
0 & 1 & \cdots & 0 & -a_{n-2} \\
\vdots & \vdots & \ddots & \vdots & \vdots \\
0 & 0 & \cdots & 1 & -a_1 \\
\end{bmatrix} \mathbf{x}(k) + \begin{bmatrix}
b_n \\ b_{n-1} \\ b_{n-2} \\ \vdots \\ b_1
\end{bmatrix} u(k)$$

$$y(k) = \begin{bmatrix} 0 & 0 & \cdots & 0 & 1 \end{bmatrix} \mathbf{x}(k) + b_0 \, u(k)$$

where $D = b_0$ (typically zero for proper systems with delay).

## State-space representation for ARX models

The `StateSpaceModelDTARXSISO` class implements an ARX model:

$$A(q^{-1}) y(k) = B(q^{-1}) q^{-nk} u(k) + e(k)$$

where:
- $A(q^{-1}) = 1 + a_1 q^{-1} + a_2 q^{-2} + \cdots + a_{na} q^{-na}$
- $B(q^{-1}) = b_1 + b_2 q^{-1} + \cdots + b_{nb} q^{-nb+1}$
- $q^{-1}$ is the backward shift operator
- $nk$ is the input delay

This translates to the difference equation:

$$y(k) = -a_1 y(k-1) - a_2 y(k-2) - \cdots - a_{na} y(k-na) + b_1 u(k-nk) + b_2 u(k-nk-1) + \cdots + b_{nb} u(k-nk-nb+1)$$

**State-space representation (Matlab/Octave `arx()` compatible):**

The model uses a minimal state representation with $n = \max(na, nb+nk)$ states.

**Example: ARX(2,2,1) model**

For $na=2$, $nb=2$, $nk=1$:
- Number of states: $n = \max(2, 2+1) = 3$
- State vector: $\mathbf{x}(k) = [y(k-1), y(k-2), u(k-1)]^T$

The state-space matrices are:

$$\mathbf{A} = \begin{bmatrix}
0 & 0 & -a_1 \\
-1 & 0 & a_2 \\
0 & 1 & 0
\end{bmatrix}, \quad
\mathbf{B} = \begin{bmatrix}
b_2 \\ b_1 \\ 0
\end{bmatrix}$$

$$\mathbf{C} = \begin{bmatrix} 0 & 0 & -1 \end{bmatrix}, \quad D = 0$$

**General form for ARX(na, nb, nk):**

For general $na$, $nb$, $nk$ with $n = \max(na, nb+nk)$:

$$\mathbf{A} = \begin{bmatrix}
0 & 0 & \cdots & 0 & c_n \\
-1 & 0 & \cdots & 0 & c_{n-1} \\
0 & 1 & \cdots & 0 & c_{n-2} \\
\vdots & \vdots & \ddots & \vdots & \vdots \\
0 & 0 & \cdots & (-1)^{n-2} & c_1
\end{bmatrix}$$

where the subdiagonal contains $[1, -1, 1, -1, \ldots]$ (alternating signs) and $c_i$ are the AR coefficients with a specific sign pattern.

The last column coefficients are:
- For $i < n-na$: $c_i = 0$ (padding)
- For $i \geq n-na$: $c_i = (-1)^{n-1-i} a_{n-i}$ (AR coefficients with alternating signs)

$$\mathbf{B} = \begin{bmatrix}
b_{nb} \\ b_{nb-1} \\ \vdots \\ b_1 \\ 0 \\ \vdots \\ 0
\end{bmatrix}, \quad
\mathbf{C} = \begin{bmatrix} 0 & 0 & \cdots & 0 & -1 \end{bmatrix}, \quad D = 0$$

**Key differences from transfer function observable canonical form:**
1. Uses alternating signs on the subdiagonal instead of all 1's
2. The C matrix extracts the negative of the last state
3. Designed for minimal state dimension when $na \neq nb+nk$
